```python
import pandas as pd
import numpy as numpy
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data using Pandas
%matplotlib inline
df = pd.read_csv('job-market.csv')
df.head(2)
```

```python
plt.figure(figsize=(15,15))
ax = sns.countplot(data=df, y = 'Location', order=df["Location"].value_counts().index)
ax.set_xlabel("Number of Jobs")
ax.set_title("Number of Jobs by Location");
```

```python
ax2 = plt.pie( df[["LowestSalary", "HighestSalary"]].value_counts(), wedgeprops=dict(width=0.5), textprops=dict(color="w"))
```
